### Simple Black-Jack  : Computer Dealer vs You !!!

```python
import random
from IPython.display import clear_output
```

```python
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
rank_to_weight_dict  = { 'Two': 2 , 'Three':3 , 'Four': 4 , 'Five':5 , 'Six':6,'Seven':7 , 'Eight':8 ,'Nine':9 , 'Ten':10,
                 'Jack':10 , 'Queen':10 ,'King':10 ,'Ace': 1}
```

```python
class Card:
    '''
    This is a card class which resembles a card from a deck of cards. It has centain properties such as 
    suit, rank and weightage. Standard __str__() fucntion has also been defined for this class for printing it out.
    '''
    def __init__(self,suit,rank):
        self.rank = rank.capitalize()
        self.suit = suit.capitalize()
        self.weightage = rank_to_weight_dict[rank.capitalize()]
    
    def __str__(self):
        return f'{self.rank}_of_{self.suit}'
```

```python
class Deck:
    
    '''
    This is Deck class ..every time you create an object of this class its like you are 
    bringing a fresh set of deck of cards with all the 52 cards present.
    
    It will have a list of all cards present in the deck @any time instant as an attribute
    and Shuffling and taking one cards out from the deck as the methods for this class object.
    '''
    
    def __init__(self):
        
        self.all_cards = []
        
        for suit in suits:
            for rank in ranks:
                new_card = Card(suit,rank)
                self.all_cards.append(new_card)
    
    #we need a function that will shuffle the deck of cards :
    def shuffle_deck(self):
        random.shuffle(self.all_cards)
        
    #we need another function to tak one card out of the deck while we dstribute it to the players 
    def take_one_card(self):
        #this pop() also reduces the length of the cards present in the deck 
        #verify the same while testing 
        return self.all_cards.pop()
```

```python
class player:
    
    '''
    This is player class. Any player participating in this game will be an object of this class. 
    
    a) It will have a name 
    b) It will have a set of cards and 
    b) It will have some balance which it can bet on.
    
    At the same time it will be capable of 
    
    a) adding cards to its collection of cards 
    b) seeing all its cards 
    c) throw a card when its turn come up 
    
    '''
    
    def __init__(self,name,balance = 0 ):
        #initially the list will be empty
        self.my_cards = []
        self.name = name
        self.cards_val = 0
        self.balance = balance
        self.ace_val = None
        
    def set_ace_val(self):
        
        global rank_to_weight_dict
        
        temp = 'Wrong'
        
        while temp not in ['Y','N','y','n']:
            temp = input('\nDo you want to set the Ace card value ?')        
            if temp == 'Y' or temp == 'y':                
                val = 0
                while int(val) not in [1,11]:
                    val = input('\nPlease enter a valid ace value(either 1 or 11) : ')
                    if val.isdigit() == False:
                        print('\nPlease enter only numeric value !!!!')
                        val = 0
                    if int(val) not in [1,11]:
                        print('\nPlease choose either 1 or 11 !!!')

                rank_to_weight_dict['Ace'] = int(val)

        
    def add_cards(self , new_card):
        self.my_cards.append(new_card)
        
    def show_my_cards(self):
        print('\n You have the below cards at present:\n')
        for index,cards in enumerate(self.my_cards,start = 1):
            print( f"{index}. {cards}")
            
    def bet_money(self):
        
        amount = 99999999999
        
        while int(amount) not in list(range(0,self.balance)):
            amount = input("how much money would you like to bet ? ")
            
            if amount.isdigit() == False:
                print('\nPlease enter a numeric value !!!')
                amount = 99999999999
                continue
            if int(amount) not in list(range(0,self.balance)):
                print('\nPlease enter a proper value as per your balance !!!')
                continue
                
        self.balance -= int(amount)
        
        return int(amount)
            
    def total_cards_val(self):
        total_val = 0 
        for cards in self.my_cards:
            total_val += rank_to_weight_dict[cards.rank]
        return total_val
    
    def decide_hit_stay(self):
        decision = 99
        while int(decision) not in [1,2]:
            decision = input('Would you like to hit or stay (Enter 1 for Hit and 2 for Stay)? ')
            
            if decision.isdigit() == False :
                print('Please enter a numeric value 1 or 2 !!!!')
                decision = 99
            if int(decision) not in [1,2]:
                print('Please enter a valid value either 1 or 2 !!!')
        
        if int(decision) == 1 :
            return 'Hit'
        else:
            return 'Stay'
```

     ----------------- All is set now , Combine the pieces in the flow and play the game :------------------


    Time to write the logic for the game:
    
    Step 1 : Create/bring a fresh deck of cards 
    Step 2 : Shuffle the deck.
    Step 3 : Decide the bet ammount.
    Step 3 : a)Take 1 cards out for you 
             b)Take one card for dealer  (You can see the value of this card.
             c)Take another card for you 
             d) Take one card for dealer !!!
    
    ----------------------------------------------Set --------------------------------------------------
    Step 4 : You will sum up the card values.
    
    step 5 : Ask Second Player to throw a card
    Step 6 : Compare and decide who wins !!!! Add 1 point to winning player.
             NOTE : If Its a draw, we keep the thrown cards in the vault , till one of them wins. 
                    winner this time gets all the cards and n points based if n sets were in the vault. 
    Step 7 : Repeat step 4-6 till one of them has no more cards with him/her to continue playing 
    
    ---------------------------------------------Game --------------------------------------------------
    Step 8 : Now decide the winner based  on the points scored by them.
    
    Step 9 : Ask the player, whether they want to play another game of War.
             If yes , Start from Step 1 else Exit() .
    

```python
def want_to_play():
    '''
    This function take input from the user whether he/she wants to continue playing the Game of Cards ??
    '''
    want_to_play = 'WRONG'

    while want_to_play.upper() not in ['Y','N','y','n','yes','no','Yes','No']:

        want_to_play = input('Hi Would you like to play the black-Jack game  (Y/N)???')

        if want_to_play.upper() not in ['Y','N','y','n','yes','no','Yes','No']:
            clear_output()
            print('\n\nPlease enter a valid input (Y/N) !!!\n\n')
    return want_to_play
```

```python
def game_set_card_withdraw(newDeck,me):
    
    global dealer_card_list
    
    #-----------------------------Take the first card and add ---------------------------------
        
    #take one card for me
    card = newDeck.take_one_card()
    #add it to my set
    me.add_cards(card)
    
    #take one card for me
    card = newDeck.take_one_card()
    #add it to my set
    dealer_card_list.append(card)
    
    #-----------------------------Take the second and add -----------------------------------
    
    #take one card for me
    card = newDeck.take_one_card()
    #add it to my set
    me.add_cards(card)
    
    #take one card for me
    card = newDeck.take_one_card()
    #add it to my set
    dealer_card_list.append(card)
    
    return newDeck,me
```

```python
def decide_winner(me,amount_at_stake):
    
    global dealer_card_list
    dealer_card_sum_val = dealer_card_sum()
    
    if me.total_cards_val() > dealer_card_sum_val:
        me.balance = me.balance + (2 * amount_at_stake)
        print(f'\n\n    Congralulations !!! You win {amount_at_stake} and Now your total balance is : {me.balance}\n\n')
        return 'Game Over'
    else:
        print(f'\n\n    Sorry !!! You Loose {amount_at_stake} and Now your total balance is : {me.balance}\n\n')
        return 'Game Over'
```

```python
def Hit(newDeck,me):
    #take one card for me
    card = newDeck.take_one_card()
    print(f'\nCard Taken out for you is : {card}')
    #add it to my set
    me.add_cards(card)
```

```python
def stay(me,newDeck):
    
    me.set_ace_val()
    
    global dealer_card_list 
    
    print(f"\nDealer's 1st card :{dealer_card_list[0]} ")
    print(f"\nDealer's 2nd card :{dealer_card_list[1]} ")
    
    dealer_card_sum_val = dealer_card_sum()
    print(f"\nDealer card sum value : {dealer_card_sum_val}")
    
    while dealer_card_sum_val < 17:
        card = newDeck.take_one_card()
        print(f'\nTaken out card for dealer is :{card} ')
        dealer_card_list.append(card)
        dealer_card_sum_val = dealer_card_sum()  
        print(f"\nUpdated Dealer card sum value : {dealer_card_sum_val}")
```

```python
def check_dealer_bust():    
    #--------------dealer busted  or not ?-----------
    dealer_card_sum_val =  dealer_card_sum()   
    if dealer_card_sum_val > 21 :
        return True
```

```python
def dealer_card_sum():
    global dealer_card_list
    dealer_card_sum_val = 0 
    
    for cards in dealer_card_list:
        dealer_card_sum_val += rank_to_weight_dict[cards.rank]
        
    return dealer_card_sum_val
```

```python
def check_bust_win(me,amount_at_stake):
    
    summed_value_of_my_cards = me.total_cards_val()
    
    if summed_value_of_my_cards == 21:
        me.balance = me.balance + (2 * amount_at_stake)
        print(f'\n\n   Congralulations !!! You win {amount_at_stake} and Now your total balance is : {me.balance}\n\n')
        return 'Game Over'
        
    if (summed_value_of_my_cards > 21):
        print(f'\n\n   Sorry !!! You busted {amount_at_stake} and Now your current balance is : {me.balance}\n\n')
        return 'Game Over'
    
    else:
        return 'Game On'
```

#### -------------------------------------------------------------Declare the Player  &  Dealer -----------------------------------------------------------------------

```python
me = player('Srijeet' , 50)
dealer_card_list = []
```

```python
play = 'Y'

while play in ['Y','y','yes','Yes'] and me.balance != 0:
    
    #bring a fresh deck of card 
    newDeck = Deck()
    
    #create fresh dealer card list 
    global dealer_card_list
    dealer_card_list.clear()
    
    me.my_cards = []
    
    #shuffle the deck
    newDeck.shuffle_deck()
    
    #-------------------------------bet the ammount ------------------------------------------
    
    amount_at_stake = me.bet_money()
        
    #-----------------------------Take the first & 2nd card and add ---------------------------------
    game_set_card_withdraw(newDeck,me)
    
    #----So now both me and dealer has two cards , Lets ee and then proceed withthe game  ---------------------------
    me.show_my_cards()
    print('\nThe dealer has the below card:\n')
    print(dealer_card_list[0])
    
    #--------------- Want to set Ace Value ?-------------------------------------------

    game_status = 'Game On'
    
    while game_status == 'Game On':
        
        me.set_ace_val()
        #---------------Take a Decision Hit or continue --------------------------------------------
        game_status = check_bust_win(me,amount_at_stake)
        
        if game_status == 'Game Over':
            break
        else:
            #decide to hit or stay
            decision = me.decide_hit_stay()
            
            if decision == 'Hit':
                Hit(newDeck,me)
                continue
            elif decision == 'Stay':
                stay(me,newDeck)
                #---------------------check whether dealer is busted or not : -------------------        
                if check_dealer_bust() == True:
                    me.balance = me.balance + (2 * amount_at_stake)
                    print(f'\n\n    Congralulations !!! You win {amount_at_stake} as Dealer Busted and Now your total balance is : {me.balance}\n\n')
                    game_status = 'Game Over'
                else:
                    game_status = decide_winner(me,amount_at_stake)
                        
    play = want_to_play()
    
    print(f'\n\n Mr. {me.name} hope you have enjoyed the Game !!! Please revisit but remember Gambling is bad !!!! ')
```
