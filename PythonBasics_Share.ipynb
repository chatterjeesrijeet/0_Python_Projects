```python
# Function to convert number into string
# Switcher is dictionary data type here
def numbers_to_strings(argument):
    switcher = {
        0: "zero",
        1: "one",
        2: "two",
    }

    # get() method of dictionary data type returns 
    # value of passed argument if it is present 
    # in dictionary otherwise second argument will
    # be assigned as default value of passed argument
    return switcher.get(argument, "nothing")

# Driver program
if __name__ == "__main__":
    argument = 8
    print (numbers_to_strings(argument))
```

###  Yeild  Vs Return Keyword 

```python
# A Simple Python program to demonstrate working
# of yield

# A generator function that yields 1 for the first time,
# 2 second time and 3 third time
def simpleGeneratorFun():
    yield 1
    yield 2
    yield 3

# Driver code to check above generator function
for value in simpleGeneratorFun(): 
    print(value)
```

The yield statement suspends function's execution and sends a value 
back to the caller, but retains enough state to enable function to 
resume where it is left off. When resumed, the function continues 
execution immediately after the last yield run. This allows its code 
to produce a series of values over time, rather than computing them 
at once and sending them back like a list.

```python
# A Python program to generate squares from 1
# to 100 using yield and therefore generator

# An infinite generator function that prints
# next square number. It starts with 1
def nextSquare():
    i = 1;

    # An Infinite loop to generate squares 
    while True:
        yield i*i                
        i += 1  # Next execution resumes 
                # from this point     

# Driver code to test above generator 
# function
for num in nextSquare():
    if num > 100:
         break    
    print(num)
```

```python
# Python code to illustrate cube of a number 
# showing difference between def() and lambda().
def cube(y):
    return y*y*y;

g = lambda x: x*x*x

print(g(7))

print(cube(7))
```

Lambda functions can be used along with built-in functions like filter(), map() and reduce().

```python
# Python code to illustrate
# filter() with lambda()
li = [5, 7, 22, 97, 54, 62, 77, 23, 73, 61]
final_list = list(filter(lambda x: (x%2 != 0) , li))
print(final_list)
```

```python
# Python code to illustrate 
# map() with lambda() 
# to get double of a list.
li = [5, 7, 22, 97, 54, 62, 77, 23, 73, 61]
final_list = list(map(lambda x: x*2 , li))
print(final_list)
```

```python
# Python code to illustrate 
# reduce() with lambda()
# to get sum of a list

from functools import reduce

li = [5, 8, 10, 20, 50, 100]
sum = reduce((lambda x, y: x + y), li)
print (sum)
```

```python
class MobilePhone:
  """This is a sample class to illustrate how Python classes work"""
  def __init__(self, name, is_android = False, screen_size = 4.3):
    self.name = name
    self.is_android = is_android
    self.screen_size = screen_size
    self.rating = -1
    
  def has_rating(self):
    return self.rating > -1
```

```python
new_phone = MobilePhone('iPhone 5s')
```

```python
type(new_phone)
```

```python
print(new_phone.name, new_phone.is_android, new_phone.screen_size)
```

```python
new_phone.screen_size = 4
```

```python
new_phone.has_rating()
```

```python
class iPhone(MobilePhone):
  def __init__(self, name):
    MobilePhone.__init__(self, name, False, 4)
    
  def __str__(self):
    return self.name + " " + str(self.is_android) + " " + str(self.screen_size)
```
